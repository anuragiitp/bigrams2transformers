```python
import torch
print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())


# Create two large tensors on the GPU
a = torch.randn(5000, 5000, device="cuda")
b = torch.randn(5000, 5000, device="cuda")

# Perform a matrix multiplication on the GPU
c = torch.matmul(a, b)

# Check memory usage after operation
print("Memory Allocated:", torch.cuda.memory_allocated() / 1024**2, "MB")
print("Memory Reserved:", torch.cuda.memory_reserved() / 1024**2, "MB")
```
